```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from skimage.io import imread
from skimage.color import rgb2lab, lab2rgb
import math 
```

# Step 1: Splitting
In this step we split the video into frames, each frame needs to be processed individually.
Therefore, we create an array with frames

```python
# Function that takes the location of the video and splits the video into frames in CIELAB Color Space
def splitFrames(videoLoc):
    cap= cv2.VideoCapture(videoLoc)
    totalframecount= int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    frames = [] # np.array(totalframecount)
    i = 0
    while(cap.isOpened()):
        ret, frameBGR = cap.read()
        if ret == False:
            break
        frameLab = cv2.cvtColor(frameBGR, cv2.COLOR_BGR2LAB)
        frameRGB = cv2.cvtColor(frameBGR, cv2.COLOR_BGR2RGB)
        frames.insert(i,frameLab)
        plt.figure(figsize=(10,5))
        plt.subplot(121), plt.imshow(frameRGB), plt.axis('off'), plt.title('RGB Frame nr: ' + str(i), size=10)
        plt.show()
        i+=1
    
    cap.release()
    cv2.destroyAllWindows()
    return frames
```

```python
# List of all the frames in CIELAB Color Space
frames = splitFrames('./preamble-2.mp4')
```

# Step 2: Finding the stripes
In this step we need to find the width of the individual stripes.
We can do this by using the preamble because we know that the preamble consists of white light // blue light // white light // blue light.
After we calculate the width of the stripes, we can divide the frame into stripes and find the color value of each individual stripe.

```python
symbols = []

frame = frames[0]
for i in range(len(frame)):
    #print(frame[i][int(len(frame[5])/2)])
    y = int(len(frame[5])/2)
    x = i
    
    lWhite = 130
    aWhite = 149
    bWhite = 91
    
    lBlue = 205
    aBlue = 115
    bBlue = 125
    
    l = frame[x][y][0]
    a = frame[x][y][1]
    b = frame[x][y][2]
    
    #pow((lwhite - l), 2) + pow((awhite - a), 2) + pow((bwhite - b), 2)
    #pow((lBlue - l), 2) + pow((aBlue - a), 2) + pow((bBlue - b), 2)
    
    euclidWhite = math.sqrt(pow((lWhite - l), 2) + pow((aWhite - a), 2) + pow((bWhite - b), 2))
    euclidBlue = math.sqrt(pow((lBlue - l), 2) + pow((aBlue - a), 2) + pow((bBlue - b), 2))
    
    if euclidWhite < euclidBlue :
        symbols.append("White")
        #print("White")
    else:
        symbols.append("Blue")
        #print("Blue")
    
    #if (abs(frame[x][y][0] - 205) < 5) or (abs(frame[x][y][1] - 115) < 5) or (abs(frame[x][y][2] - 125) < 5):
    #    symbols.append("Blue")
    #    print("Blue")
    #if (abs(frame[x][y][0] - 149) < 5) or (abs(frame[x][y][1] - 130) < 5) or (abs(frame[x][y][2] - 91) < 5):
    #    symbols.append("White")
    #    print("White")
    #else:
        #print("Unknown")
        #symbols.append("Unknown")
```

```python
countedSym = []
t = 0
countedSym.append((symbols[0], 1))

for i in range(1, len(symbols)):
    prev = countedSym.pop()
    
    if (symbols[i] == "White") and (prev[0] == "White"):
        count = prev[1] +1
        upd = (symbols[i], count)
        countedSym.append(upd)
        #countedSym[t] = countedSym[t][1] + 1
        
    if (symbols[i] == "Blue") and (prev[0] == "Blue"):
        countedSym.append(("Blue", prev[1] + 1))
        #countedSym[t][1] = countedSym[t][1] + 1
        
    if (symbols[i] == "Unknown") and (prev[0] == "Unknown"):
        countedSym.append(("Unknown", prev[1] + 1))
        #countedSym[t][1] = countedSym[t][1] + 1 
        
    if (symbols[i] == "White") and (prev[0] == "Blue" or prev[0] == "Unknown"):
        countedSym.append(prev)
        t = t + 1
        countedSym.append(("White", 1))
        
    if (symbols[i] == "Blue") and (prev[0] == "White" or prev[0] == "Unknown"):
        countedSym.append(prev)
        t = t + 1
        countedSym.append(("Blue", 1))
        
    if (symbols[i] == "Unknown") and (prev[0] == "White" or prev[0] == "Blue"):
        countedSym.append(prev)
        t = t + 1
        countedSym.append(("Unknown", 1))  
```

```python
upCountedSym = []
for i in range (len(countedSym)):
    if countedSym[i][1] > 0:
        upCountedSym.append(countedSym[i])
```
