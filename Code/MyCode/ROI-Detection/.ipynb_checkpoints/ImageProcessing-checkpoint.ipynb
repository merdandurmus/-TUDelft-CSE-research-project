```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from skimage.io import imread
from skimage.color import rgb2lab, lab2rgb
import math 
```

# Step 1: Splitting
In this step we split the video into frames, each frame needs to be processed individually.
Therefore, we create an array with frames
USE CIELAB space because then its better

```python
# Function that takes the location of the video and splits the video into frames in CIELAB Color Space
def splitFrames(videoLoc):
    cap= cv2.VideoCapture(videoLoc)
    totalframecount= int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    frames = [] # np.array(totalframecount)
    i = 0
    while(cap.isOpened() and i < 50):
        ret, frameBGR = cap.read()
        if ret == False:
            break
        #frameLab = cv2.cvtColor(cv2.rotate(frameBGR, cv2.ROTATE_90_CLOCKWISE), cv2.COLOR_BGR2LAB)
        #frameRGB = cv2.cvtColor(cv2.rotate(frameBGR, cv2.ROTATE_90_CLOCKWISE), cv2.COLOR_BGR2RGB)
        frameLab = cv2.cvtColor(frameBGR, cv2.COLOR_BGR2LAB)
        frameRGB = cv2.cvtColor(frameBGR, cv2.COLOR_BGR2RGB)
        frames.insert(i,frameLab)
        plt.figure(figsize=(10,5))
        plt.subplot(121), plt.imshow(frameRGB), plt.axis('off'), plt.title('RGB Frame nr: ' + str(i), size=10)
        plt.show()
        i+=1
    
    cap.release()
    cv2.destroyAllWindows()
    return frames
```

```python
# List of all the frames in CIELAB Color Space
frames = splitFrames('./finalcsk33.mp4')
```

We can do this by using the preamble because we know that the preamble consists of white light // blue light // white light // blue light.
After we calculate the width of the stripes, we can divide the frame into stripes and find the color value of each individual stripe.

```python
sym0 = [149, 60]
sym1 = [187, 84]
sym2 = [193, 93]
sym3 = [152, 63]
sym4 = [98, 102]
sym5 = [104, 206]
sym6 = [103, 102]
sym7 = [98, 102]

colors = [sym0, sym1, sym2, sym3, sym4, sym5, sym6, sym7]

colorToSymbol = []
def findCorrespondingSymbol(stripes):
    for i in range(len(stripes)):
        a = stripes[i][0]
        b = stripes[i][1]
        
        tempEuclid = 9999999
        tempInd = -1
        
        for j in range (len(colors)):
            ca = colors[j][0]
            cb = colors[j][1]
            
            euclid = math.sqrt(pow((ca - a),2) + pow((cb - b),2))
             
            #print("For stripe: " + str(i) + " and synchronized symbol: " + str(j) + " ,the euclidean distance is: " + str(euclid))
            
            if (euclid < tempEuclid):
                tempEuclid = euclid
                tempInd = j
        
        colorToSymbol.append(tempInd)
                

hardCodeStripes = [[120, 90], [123, 90], [123, 116], [113, 193], [159, 113], [155, 114], [156, 72]]
#findCorrespondingSymbol(hardCodeStripes)
findCorrespondingSymbol(symbols)
colorToSymbol
```

```python
def toBit(b):
    if b == 0:
        return "000"
    if b == 1:
        return "001"
    if b == 2:
        return "010"
    if b == 3:
        return "011"
    if b == 4:
        return "100"
    if b == 5:
        return "101"
    if b == 6:
        return "110"
    if b == 7:
        return "111"

bits = ""
for i in range(len(colorToSymbol)):
    bits += (toBit(colorToSymbol[i]))
  
bits
#toBit(colorToSymbol)
```
